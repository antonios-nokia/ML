```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib  # For saving and loading models

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species']) 
```

```python
X = df.drop('species', axis=1)  # Features
y = df['species']  # Target
```

```python
print(y)
```

```python
# Normalize feature values using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Save the trained model, scaler, and label encoder
joblib.dump(model, "iris_logistic_model.pkl")
joblib.dump(scaler, "iris_scaler.pkl")
joblib.dump(label_encoder, "iris_label_encoder.pkl")

print("Model, scaler, and label encoder saved successfully.")
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Function to predict new Iris flower species
def predict_new_iris(input_data):
    # Load saved model, scaler, and label encoder
    model = joblib.load("iris_logistic_model.pkl")
    scaler = joblib.load("iris_scaler.pkl")
    label_encoder = joblib.load("iris_label_encoder.pkl")

    # Convert input to the correct format and scale it
    input_scaled = scaler.transform(np.array(input_data).reshape(1, -1))

    # Predict species
    prediction = model.predict(input_scaled)
    
    # Convert numeric prediction back to species name
    species_name = label_encoder.inverse_transform(prediction)[0]

    return species_name
```

```python
# Example: Predict a new flower species using the first row of the dataset
new_iris = X.iloc[0].tolist()
print("Predicted Iris Species:", predict_new_iris(new_iris))
```
